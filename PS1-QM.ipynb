### Exercise 1.1.
Approximate $f(x) = x^.321$ with a Taylor series around $\bar x = 1$. Compare your approximation over the domain (0,4). Compare when you use up to 1, 2, 5 and 20 order approximations. Discuss your results.

```python
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt

# Define the variable and the function to approximate
x = sy.Symbol('x')
f = x**.321


# Factorial function
def factorial(n):
    if n <= 0:
        return 1
    else:
        return n*factorial(n-1)

# Taylor approximation at x0 of the function 'function'
def taylor(function,x0,n):
    i = 0
    p = 0
    while i <= n:
        p = p + (function.diff(x,i).subs(x,x0))/(factorial(i))*(x-x0)**i
        i += 1
    return p

# Plot results
def plot():
    x_lims = [0,4]
    x1 = np.linspace(x_lims[0],x_lims[1],800)
    y1 = []
    # Use it up to 1, 2, 5 and 20 order approximations. 
    order = [1, 2, 5, 20]
    for j in order:
        func = taylor(f,1,j)
        for k in x1:
            y1.append(func.subs(x,k))
        plt.plot(x1,y1,label='order '+str(j))
        y1 = []
    # Plot the function to approximate
    plt.plot(x1,x1**.321,label='CD')
    plt.xlim(x_lims)
    plt.ylim([0,4])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.title('Graph 1: Taylor approximation of CD function')
    plt.show()

plot()
```

### Exercise 1.2.
Approximate the ramp function $f(x) = \frac{x+|x|}{2}$ with a Taylor series around $\bar x = 2$. Compare your approximation over the domain (-2,6). Compare when you use up to 1, 2, 5 and 20 order approximations. Discuss your results.


```python
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
import math


# Define the variable and the function to approximate
x = sy.Symbol('x', real=True)
f = 0.5*(x + abs(x))


# Factorial function
def factorial(n):
    if n <= 0:
        return 1
    else:
        return n*factorial(n-1)

# Taylor approximation at x0 of the function 'function'
def taylor(function,x0,n):
    i = 0
    p = 0
    while i <= n:
        p = p + (function.diff(x,i).subs(x,x0))/(factorial(i))*(x-x0)**i
        i += 1
    return p

# Plot results
def plot():
    x_lims = [-2,6]
    x1 = np.linspace(x_lims[0],x_lims[1],800)
    y1 = []
    # Use it up to 1, 2, 5 and 20 order approximations. 
    order = [1, 2, 5, 20]
    for j in order:
        func = taylor(f,2,j)
        for k in x1:
            y1.append(func.subs(x,k))
        plt.plot(x1,y1,label='order '+str(j))
        y1 = []
    # Plot the function to approximate
    plt.plot(x1, 0.5*(x1 + abs(x1)),label='ramp function')
    plt.xlim(x_lims)
    plt.ylim([-2,6])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.title('Graph 2: Taylor approximation of the Ramp Function')
    plt.show()

plot()
```

The Ramp function (the analytical definition we use here is just one of the possible definitions) is a unary real function that looks like a ramp. The Taylor Series approximations fit perfectly the actual value over the non-negative domain. However, the image of the ramp function takes 0 for negative inputs; since all Taylor approximations become linear approximations (no matter the approximation order) all them are not able to capture the kink at $x=0$ (singular point).

```python
# RUNGE
# Import packages we need.
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
import numpy as np

# Define inputs and the function we want to approximate. 
domain = np.linspace(-1, 1, 100)
def f(x):
    f = 1/(1+25*(x**2))
    return f


#TWO METHODS of getting the iteration.
#METHOD 1: built-in functions
x = np.linspace(-1, 1, 10) #Define the number of nodes (for instance, 10)
p=np.polyfit(x,f(x),3) # Get the coefficients
y1=np.polyval(p,domain) #Evaluate the polynomial of the previous coefficients over the domain. 
p2=np.polyfit(x,f(x),5)
y2=np.polyval(p2,domain)



#METHOD 2: manual programming. We build a base of monomials (up to order 10) to approximate the function. 
x = np.asarray(np.linspace(-1,1,10))
y = 1/(1+25*(x**2))

# Get matrix of exponents of x values => A
A = np.zeros([10, 10])
for i in range(10):
    A[::,i] = np.power(x.T,i)
b = y

# Solve Ax=b linear eq. system to get where x denotes coeffs of polynomial in reverse order
s = np.linalg.solve(A, b)
# Flip polynomial coeffs
s = np.flip(s,axis=0)

# Evaluate polynomial at the domain
y3 = np.polyval(s, domain)


#Approximation errors
error1=abs(f(domain)-y1)
error2=abs(f(domain)-y2)
error3=abs(f(domain)-y3)

#Plot everything together
plt.figure(1)

plt.subplot(121)
plt.plot(domain, f(domain), label='Runge function', lw=3)
plt.plot(domain, y1, 'g', label='cubic')
plt.plot(domain, y2,'r', label='deg 5')
plt.plot(domain, y3, label= 'deg 10')
plt.legend(['Runge', 'cubic', 'deg 5', 'deg 10'], loc='best')
plt.title('Graph 3: Runge Function-I')

plt.subplot(122)
plt.plot(domain, error1,'g', label='cubic')
plt.plot(domain, error2,'r', label='deg 5')
plt.plot(domain, error3, color='orange',label='deg 10')
plt.legend(loc='best')
plt.ylim([0,0.55])
plt.title('Graph 4: Runge approximation errors')
plt.subplots_adjust(top=0.9, bottom=0.1, left=0, right=1.5, hspace=0.5, wspace=0.5)
```

```python
#PROBABILITY FUNCTION
#Import packages
import numpy as np
import matplotlib.pyplot as plt
from math import cos

#Combination 1
#Define the function we want to interpolate:
domain = np.linspace(0, 10, 1000)
def f(x):
    f =np.exp(-x)/(5+(0.01*np.exp(-x)))
    return f

#Get the Chebyshev nodes (10 nodes) 
xl=[]
for i in range(10):
    q=cos((((2*i)-1)/20)*np.pi)
    xl.append(q)

x=np.asarray(xl)

#Interpolation of polinomials using a Chebyshev way. 
p=np.polynomial.chebyshev.chebfit(x,f(x),3)
y1=np.polynomial.chebyshev.chebval(domain,p)
p1=np.polynomial.chebyshev.chebfit(x,f(x),5)
y2=np.polynomial.chebyshev.chebval(domain, p1)
p2=np.polynomial.chebyshev.chebfit(x,f(x),8)
y3=np.polynomial.chebyshev.chebval(domain, p2) 

#Approximation errors. 
error1=abs(f(domain)-y1)
error2=abs(f(domain)-y2)
error3=abs(f(domain)-y3)
                    
#Combination 2
#Define the function we want to interpolate:
domain = np.linspace(0, 10, 1000)
def g(x):
    g = np.exp(-x)/(4+(0.01*np.exp(-x)))
    return g

#Get the Chebyshev nodes (20 nodes) 
xl=[]
for i in range(10):
    q=cos((((2*i)-1)/40)*np.pi)
    xl.append(q)

x=np.asarray(xl)

#Interpolation of polinomials using a Chebyshev way. 
p=np.polynomial.chebyshev.chebfit(x,g(x),3)
y11=np.polynomial.chebyshev.chebval(domain,p)
p1=np.polynomial.chebyshev.chebfit(x,g(x),5)
y21=np.polynomial.chebyshev.chebval(domain, p1)
p2=np.polynomial.chebyshev.chebfit(x,g(x),10)
y31=np.polynomial.chebyshev.chebval(domain, p2) 

#Approximation errors. 
error11=abs(g(domain)-y1)
error21=abs(g(domain)-y2)
error31=abs(g(domain)-y3)

#Plot
plt.figure(1)

plt.subplot(121)
plt.plot(domain,f(domain), label='Probability 1', lw=3)
plt.plot(domain,g(domain), label='Probability 2', lw=2)
plt.plot(domain,y1, label='cubic')
plt.plot(domain,y2, label='deg 5')
plt.plot(domain,y3, label = 'deg 10')
plt.plot(domain,y11, label='cubic.1')
plt.plot(domain,y21, label='deg 5.1')
plt.plot(domain,y31, label = 'deg 10.1')
plt.ylim([0,0.3])
plt.xlim([0,10])
plt.legend(loc=1)
plt.title('Graph 21: Probability function approximation')

plt.subplot(122)

plt.plot(domain,error1, label='cubic')
plt.plot(domain,error2, label='deg 5')
plt.plot(domain,error3, label= 'deg 10')
plt.plot(domain,error1, label='cubic.1')
plt.plot(domain,error2, label='deg 5.1')
plt.plot(domain,error3, label= 'deg 10.1')
plt.legend(loc='best')
plt.ylim([0,0.4])
plt.xlim([0,10])
plt.title('Graph 22: Probability approx errors')
plt.subplots_adjust(top=0.9, bottom=0.1, left=0, right=1.5, hspace=0.5, wspace=0.5)
```

Approximate f(k,h) using a 2-dimensional Chebyshev regression algorithm. Fix the number of nodes to be 20 and try Cheby polynomials that go from degree 3 to 15. For each case, show the associated approximation errors (vertical axis) in the (k, h) space. Use $\sigma$=0.25

```python
# We thank Elena, Weimin and Mridula for the code for implementing the 2D Cheby algorithm. 

# 2D CHEBY ALGORITHM. CES FUNCTION WITH SIGMA=0.25
import numpy as np
from math import cos
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm

# Define the CES function. 

def f(alpha,rho,k,h):
	return np.power((1-alpha)*k**rho + alpha*h**rho,1/rho)

alpha = 0.5
sigma = 0.25
rho= (sigma -1)/sigma

# Chebyshev Approximation Algorithm in R^2. ##
# STEP 1. Get the Chebyshev nodes (20 nodes) in [-1,1].
w=[]
for i in range(20):
    q=-cos((((2*i)-1)/40)*np.pi)
    w.append(q)

w=np.asarray(w) 

# STEP 2. Adjust nodes to [a, b] and [c, d] intervals, where [a,b] is the domain of k and [c, d] the domain of h.
k_nodes=[]
for j in w:
    r= (j+1)*5
    k_nodes.append(r)

k=np.asarray(k_nodes)

# Since the domain of the state variables is the same, the nodes are the same. 
l = k

#STEP 3. Evaluate the function at the nodes. 
K,L=np.meshgrid(k,l)
approx=f(alpha,rho,K,L)

#STEP 4. Get the coefficients. 
#Step 4.1. Chebyshev polynomials of degree d evaluated at the vector x. 
def T(d,x): 
	poly = []
	poly.append(np.ones(len(x)))
	poly.append(x)
	for i in range (1,d):
		p = 2* x *poly[i-1] - poly[i-2]
		poly.append(p)
	poly_mat = np.matrix(poly[d])
	return poly_mat 

def coeff(z,y_nodes,d):
    theta=np.empty((d+1)*(d+1))
    theta.shape = (d+1,d+1)
    for i in range(d+1):
        for j in range(d+1):
            theta[i,j] = (np.sum(np.array(y_nodes)*np.array((np.dot(T(i,z).T,T(j,z)))))
                          /np.array((T(i,z)*T(i,z).T)*(T(j,z)*T(j,z).T)))
    return theta

theta = coeff(w,approx,d=2)

#STEP 5. Approximation for f(x) at the selected interval
a=0
b=10

def f_approx(x,y, theta, d):
    f = []
    in1 = (2*(x/10))-1
    in2 = (2*(y/10))-1
    for u in range(d):
        for v in range(d):
                f.append(np.array(theta[u,v])*np.array((np.dot(T(u,in1).T,T(v,in2)))))
    f_sum = sum(f)
    return f_sum

#ERRORS
# True
domain=np.linspace(0.01,10,20)
X,Y=np.meshgrid(domain,domain)
Z = f(alpha,rho,X,Y)

#Approximation
Z1 = f_approx(domain,domain,theta,d=2)

e1=abs(Z-Z1)

#PLOT
fig = plt.figure(figsize=plt.figaspect(0.25))
fig.suptitle('Graph 23: Chebyshev approx order 3 for σ=0.25', fontsize=16)

ax = fig.add_subplot(131, projection='3d')
ax.set_xlabel('Capital')
ax.set_ylabel('Labor')
ax.set_zlabel('Output')
ax.contour3D(X, Y, Z, 100)
ax.set_title('CES function')

ax = fig.add_subplot(132, projection='3d')
ax.set_xlabel('Capital')
ax.set_ylabel('Labor')
ax.set_zlabel('Output')
ax.contour3D(X,Y, Z1, 100)
ax.set_title('Chebyshev approximation order 3')

ax = fig.add_subplot(133, projection='3d')
ax.set_xlabel('Capital')
ax.set_ylabel('Labor')
ax.set_zlabel('Output')
ax.contour3D(X, Y, e1, 100)
ax.set_title('Approximation errors')
```
